EJERCICIO 01:

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------
# Ejercicio 1a
# Sistema: 5x + 2y = 1 ; -3x + 3y = 5
# ----------------------
A1a = np.array([[5, 2], [-3, 3]])
B1a = np.array([1, 5])
sol1a = np.linalg.solve(A1a, B1a)

x_sol, y_sol = sol1a
print(f"Ejercicio 1a - Solución: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

x_vals = np.linspace(-5, 5, 200)
y1a = (1 - 5*x_vals)/2
y2a = (5 + 3*x_vals)/3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1a, label="5x + 2y = 1")
plt.plot(x_vals, y2a, label="-3x + 3y = 5")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 1a")
plt.show()


# ----------------------
# Ejercicio 1b
# Sistema: 2x + y = 6 ; 4x + 3y = 14
# ----------------------
A1b = np.array([[2, 1], [4, 3]])
B1b = np.array([6, 14])
sol1b = np.linalg.solve(A1b, B1b)

x_sol, y_sol = sol1b
print(f"Ejercicio 1b - Solución: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

y1b = 6 - 2*x_vals
y2b = (14 - 4*x_vals)/3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1b, label="2x + y = 6")
plt.plot(x_vals, y2b, label="4x + 3y = 14")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 1b")
plt.show()
```

EJERCICIO 04:

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------
# Ejercicio 4a
# Sistema: -2x + 3y = 14 ; 3x - y = -14
# ----------------------
A4a = np.array([[-2, 3], [3, -1]])
B4a = np.array([14, -14])
sol4a = np.linalg.solve(A4a, B4a)

x_sol, y_sol = sol4a
print(f"Ejercicio 4a - Solución: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

x_vals = np.linspace(-10, 10, 200)
y1a = (14 + 2*x_vals)/3
y2a = 3*x_vals + 14

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1a, label="-2x + 3y = 14")
plt.plot(x_vals, y2a, label="3x - y = -14")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 4a")
plt.show()


# ----------------------
# Ejercicio 4b
# Sistema: 2x + 3y = 2 ; -6x + 12y = 1
# ----------------------
A4b = np.array([[2, 3], [-6, 12]])
B4b = np.array([2, 1])

try:
    sol4b = np.linalg.solve(A4b, B4b)
    x_sol, y_sol = sol4b
    print(f"Ejercicio 4b - Solución: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")
except np.linalg.LinAlgError:
    print("Ejercicio 4b - El sistema no tiene solución única (es incompatible o tiene infinitas soluciones).")
    sol4b = None

y1b = (2 - 2*x_vals)/3
y2b = (1 + 6*x_vals)/12

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1b, label="2x + 3y = 2")
plt.plot(x_vals, y2b, label="-6x + 12y = 1")
if sol4b is not None:
    plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 4b")
plt.show()
```

```python
# Ejercicio 20
# a) De los pares: (0,10), (3,19/2), (-1,-4), (0,2/5), (-1/2,7)
#    ¿cuáles son soluciones de -3x + (1/2) y = 5 ?
# b) Representa gráficamente la recta -3x + (1/2) y = 5.
# c) ¿Qué relación hay entre los puntos de la recta y las soluciones?

import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction

# Lista de pares (usar Fraction para conservar exactitud si se desea)
pairs = [
    (0, 10),
    (Fraction(3,1), Fraction(19,2)),   # (3, 19/2)
    (-1, -4),
    (0, Fraction(2,5)),
    (Fraction(-1,2), 7)
]

# Ecuación: -3x + (1/2)y = 5  -> despejar y: (1/2)y = 5 + 3x  -> y = 2*(5 + 3x) = 10 + 6x
def lhs_value(x, y):
    return -3*x + Fraction(1,2)*y

print("Comprobación de pares respecto a -3x + (1/2)y = 5:\n")
solutions_found = []
for p in pairs:
    x_p, y_p = p
    val = lhs_value(x_p, y_p)
    is_solution = (val == 5)
    print(f" Punto {p}: LHS = {val} -> solución: {is_solution}")
    if is_solution:
        solutions_found.append(p)

print("\nPuntos que son soluciones:", solutions_found)

# Graficar la recta y marcar puntos
x_vals = np.linspace(-6, 6, 400)
y_vals = 10 + 6*x_vals  # como mostramos arriba: y = 10 + 6x

plt.figure(figsize=(7,6))
plt.plot(x_vals, y_vals, label=r"-3x + (1/2)y = 5  (=> y = 10 + 6x)")

# marcar todos los pares; si cumplen, color verde, sino color red
for p in pairs:
    xp = float(p[0])
    yp = float(p[1])
    color = 'green' if p in solutions_found else 'red'
    plt.scatter(xp, yp, color=color, zorder=5)
    plt.text(xp+0.1, yp+0.1, str(p), fontsize=9)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Ejercicio 20 — Recta y comprobación de pares")
plt.show()

# Respuesta (c)
print("\nRespuesta (c):")
print(" Los puntos que están sobre la recta (es decir, los que al sustituir en la ecuación dan igualdad) son exactamente las soluciones.")
print(" Cada punto (x,y) que yace en la recta satisface -3x + (1/2)y = 5; los demás no.")
```

```python
# Ejercicio 21
# Sistema:
#  -x + y = 5
# -2x + 2y = 2
#
# Averigua cuántas soluciones tiene el sistema representando las dos rectas.

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x, y = sp.symbols('x y')

# Ecuaciones (igualadas a 0)
ec1 = -x + y - 5
ec2 = -2*x + 2*y - 2

# Resolver simbólicamente (si es posible)
try:
    sol21 = sp.solve([ec1, ec2], (x, y))
    has_solution = True
except Exception as e:
    sol21 = None
    has_solution = False

print("🔹 Ejercicio 21")
print("Ecuaciones:")
print("  -x + y = 5")
print(" -2x + 2y = 2")

# Analizar dependencia / compatibilidad usando matrices
A = sp.Matrix([[-1, 1], [-2, 2]])
B = sp.Matrix([5, 2])
rankA = A.rank()
rankAug = A.row_join(B).rank()

print(f"Rango(A) = {rankA}, Rango(A|b) = {rankAug}")

if rankA == rankAug:
    if rankA == 2:
        print("-> Sistema compatible determinado: una única solución.")
    else:
        print("-> Sistema compatible indeterminado: infinitas soluciones (las rectas coinciden).")
else:
    print("-> Sistema incompatible: no tiene soluciones (rectas paralelas).")

# Si hay solución única, mostrarla
if isinstance(sol21, dict) and sol21:
    print("Solución exacta:", sol21)
elif sol21 == [] or sol21 is None:
    print("No hay solución única simbólica directa (ver rangos).")

# Graficar las rectas
x_vals = np.linspace(-5, 10, 400)
y1 = (x_vals + 5)          # de -x + y = 5 => y = x + 5
y2 = (2*x_vals + 2)/2      # de -2x + 2y = 2 => 2y = 2x + 2 => y = x + 1

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1, label="-x + y = 5  -> y = x + 5")
plt.plot(x_vals, y2, label="-2x + 2y = 2 -> y = x + 1")

# Si hay solución (intersección) calcúlela y márcala
if rankA == rankAug and rankA == 2:
    sol_num = sp.solve([ec1, ec2], (x, y))
    px, py = float(sol_num[x]), float(sol_num[y])
    plt.scatter(px, py, color='red', zorder=5, label=f"Intersección ({px:.3f}, {py:.3f})")

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 21 — Rectas")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()

# Conclusión textual
if rankA == rankAug:
    if rankA == 2:
        print("Conclusión: el sistema tiene UNA solución (rectas se cruzan en un punto).")
    else:
        print("Conclusión: infinitas soluciones (rectas coinciden).")
else:
    print("Conclusión: no tiene soluciones (rectas paralelas).")
```

```python
# Problema 8
# Perímetro = 19. Cada lado igual excede en 2 cm al doble del lado desigual.
# Sea base = b (lado desigual), cada lado igual = 2*b + 2
# Perímetro: b + 2*(2b + 2) = 19 -> 5b + 4 = 19 -> 5b = 15 -> b = 3

import matplotlib.pyplot as plt

b = 3.0
lado_igual = 2*b + 2
print("Problema 8 - Resultado")
print(f"  Lado desigual (base) = {b} cm")
print(f"  Lados iguales = {lado_igual} cm (cada uno)")
print(f"  Comprobación perímetro = {b + 2*lado_igual} cm")

# Dibujo aproximado del triángulo (no a escala perfecta en ángulos)
# colocamos la base horizontal de longitud b y los otros dos vértices a cierta altura
base = b
left = (0,0)
right = (base, 0)
# calcular una altura aproximada por ley de cosenos para visualización:
# tomaremos el triángulo con lados (lado_igual, lado_igual, base)
import math
a = lado_igual
c = base
# semiperímetro y área via Herón para altura:
s = (a + a + c)/2
area = math.sqrt(max(s*(s-a)*(s-a)*(s-c), 0.0))
h = 2*area / c if c!=0 else 0.0
top = (base/2, h)
xs = [left[0], right[0], top[0], left[0]]
ys = [left[1], right[1], top[1], left[1]]
plt.figure(figsize=(5,5))
plt.plot(xs, ys, '-o')
plt.text(left[0]-0.1, left[1]-0.1, f"({0:.1f},0)")
plt.text(right[0]+0.05, right[1]-0.05, f"({base:.1f},0)")
plt.title("Problema 8 — Triángulo Isósceles (ilustración)")
plt.axis('equal'); plt.grid(True); plt.show()
```

**Problema nº 10**

```python
# Problema 10
# La suma de las tres cifras de un número capicúa es 12.
# La cifra de las decenas excede en 4 unidades al doble de la cifra de las centenas.
# Halla dicho número (capicúa de 3 cifras => forma: a b a).

import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

a, b = sp.symbols('a b', integer=True)

# Interpretación:
# 2a + b = 12
# b = 2*a + 4
eq1 = sp.Eq(2*a + b, 12)
eq2 = sp.Eq(b, 2*a + 4)

sol_list = sp.solve((eq1, eq2), (a, b), dict=True)
sol = sol_list[0]

a_val = int(sol[a])
b_val = int(sol[b])
numero = 100*a_val + 10*b_val + a_val

print("Problema 10 - Resultado")
print(f"  Centenas (a) = {a_val}, Decenas (b) = {b_val}, Unidades = {a_val}")
print(f"  Número capicúa = {numero}")

# Gráfico simple: barras con las cifras
plt.figure(figsize=(5,3))
plt.bar(['centenas (a)', 'decenas (b)', 'unidades (a)'], [a_val, b_val, a_val], color=['#4c72b0','#55a868','#4c72b0'])
plt.title(f"Problema 10 — Cifras del número {numero}")
plt.ylim(0,10)
plt.grid(axis='y')
plt.show()
```

**Problema nº 11**

```python
# Problema 11
# Perímetro de rectángulo = 22 cm. Base = altura + 5 cm.
# Plantear sistema y hallar dimensiones.

import sympy as sp
import matplotlib.pyplot as plt

B, H = sp.symbols('B H', real=True)

# Interpretación:
# 2*(B + H) = 22  => B + H = 11
# B = H + 5
eq1 = sp.Eq(B + H, 11)
eq2 = sp.Eq(B, H + 5)

sol = sp.solve((eq1, eq2), (B, H), dict=True)[0]
B_val = float(sol[B])
H_val = float(sol[H])

print("Problema 11 - Resultado")
print(f"  Base = {B_val:.2f} cm")
print(f"  Altura = {H_val:.2f} cm")

# Dibujo del rectángulo
plt.figure(figsize=(5,3))
rect_x = [0, B_val, B_val, 0, 0]
rect_y = [0, 0, H_val, H_val, 0]
plt.plot(rect_x, rect_y, '-o')
plt.title("Problema 11 — Rectángulo (dimensiones)")
plt.text(B_val/2, -0.3, f"Base = {B_val:.2f} cm", ha='center')
plt.text(-0.7, H_val/2, f"Altura = {H_val:.2f} cm", va='center', rotation=90)
plt.axis('equal'); plt.grid(True); plt.show()
```

**Problema nº 12**

```python
# Problema 12
# Mezclamos dos líquidos:
# Precio1 = 0.94 €/L, Precio2 = 0.86 €/L
# Obtenemos 40 L de mezcla a 0.89 €/L.
# ¿Cuántos litros de cada tipo usamos?
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x, y = sp.symbols('x y', real=True)

# Interpretación:
# x + y = 40
# 0.94*x + 0.86*y = 0.89 * 40
eq1 = sp.Eq(x + y, 40)
eq2 = sp.Eq(sp.Rational(94,100)*x + sp.Rational(86,100)*y, sp.Rational(89,100)*40)

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 12 - Resultado")
print(f"  Litros del primer líquido (0.94 €/L): {x_val:.2f} L")
print(f"  Litros del segundo líquido (0.86 €/L): {y_val:.2f} L")
print(f"  Comprobación precio medio = {(0.94*x_val + 0.86*y_val)/40:.3f} €/L")

# Gráfico: pastel de proporción
plt.figure(figsize=(5,4))
plt.pie([x_val, y_val], labels=[f"0.94 €/L\n{x_val:.1f} L", f"0.86 €/L\n{y_val:.1f} L"], autopct='%1.1f%%', colors=['#66c2a5','#fc8d62'])
plt.title("Problema 12 — Proporción de la mezcla (40 L)")
plt.show()
```

**Problema nº 13**

```python
# Problema 13
# "El doble de un número más la mitad de otro suman 7;
#  si sumamos 7 al primero, obtenemos el quíntuple del otro."
# Plantea y resuelve el sistema.

import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x, y = sp.symbols('x y', real=True)

# Interpretación:
# 2x + (1/2) y = 7
# x + 7 = 5*y
eq1 = sp.Eq(2*x + sp.Rational(1,2)*y, 7)
eq2 = sp.Eq(x + 7, 5*y)

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 13 - Resultado")
print(f"  Primer número x = {x_val:.3f}")
print(f"  Segundo número y = {y_val:.3f}")

# Graficación: las dos rectas en el plano (x,y) y su intersección
x_vals = np.linspace(x_val-5, x_val+5, 400)
# despejar y de cada ecuación para graficar
y_from_eq1 = [float((7 - 2*xx) * 2) for xx in x_vals]     # from eq1: (1/2)y = 7 - 2x => y = 2*(7 - 2x)
y_from_eq2 = [float((xx + 7)/5) for xx in x_vals]         # eq2: y = (x + 7)/5

plt.figure(figsize=(6,5))
plt.plot(x_vals, y_from_eq1, label="2x + 1/2 y = 7  -> y = 2(7-2x)")
plt.plot(x_vals, y_from_eq2, label="x + 7 = 5y       -> y = (x+7)/5")
plt.scatter([x_val], [y_val], color='red', label=f"Solución ({x_val:.3f},{y_val:.3f})")
plt.xlabel("x"); plt.ylabel("y")
plt.legend(); plt.grid(True); plt.title("Problema 13 — Intersección de condiciones")
plt.show()
```

**Problema nº 14**

```python
# Problema 14
# Dos de los ángulos de un triángulo suman 122°.
# El tercero excede en 4 grados al menor de los otros dos.
# ¿Cuánto miden los ángulos?

import math
import matplotlib.pyplot as plt

# Si los dos suman 122°, el tercero c = 180 - 122 = 58°
c = 58
# El tercero excede en 4 al menor de los otros dos => menor = c - 4 = 54
menor = c - 4
mayor = 122 - menor  # ya que menor + mayor = 122
print("Problema 14 - Resultado")
print(f"  Ángulos: {menor}°, {mayor}°, {c}°")

# Dibujo (ilustrativo)
# Triángulo con ángulos (menor, mayor, c)
plt.figure(figsize=(5,5))
plt.text(0.1,0.8, f"Ángulos: {menor}°, {mayor}°, {c}°", fontsize=12)
plt.axis('off')
plt.title("Problema 14 — Ángulos del triángulo")
plt.show()
```

**Problema nº 15**

```python
# Problema 15
# Inversión 1: beneficio 5%   -> 0.05 * x
# Inversión 2: beneficio 3.5% -> 0.035 * y
# Se invirtió en total 10000 €: x + y = 10000
# Los beneficios de la primera inversión superan en 300 € a los de la segunda:
# 0.05 x - 0.035 y = 300
# Hallar cuánto se invirtió en cada producto.

import sympy as sp
import matplotlib.pyplot as plt

x, y = sp.symbols('x y', real=True)

eq1 = sp.Eq(x + y, 10000)
eq2 = sp.Eq(sp.Rational(5,100)*x - sp.Rational(35,1000)*y, 300)  # 0.05x - 0.035y = 300

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 15 - Resultado")
print(f"  Inversión en primer producto (5%)  = {x_val:.2f} €")
print(f"  Inversión en segundo producto (3.5%)= {y_val:.2f} €")
print(f"  Beneficio 1 = {0.05*x_val:.2f} €, Beneficio 2 = {0.035*y_val:.2f} €")
print(f"  Diferencia de beneficios = {0.05*x_val - 0.035*y_val:.2f} € (comprobación)")

# Gráfico: barras comparando las inversiones y beneficios
plt.figure(figsize=(7,4))
plt.subplot(1,2,1)
plt.bar(['Inv1 (5%)','Inv2 (3.5%)'], [x_val, y_val], color=['#4c72b0','#dd8452'])
plt.title("Inversiones")
plt.ylabel("Euros")
plt.subplot(1,2,2)
plt.bar(['Beneficio1','Beneficio2'], [0.05*x_val, 0.035*y_val], color=['#4c72b0','#dd8452'])
plt.title("Beneficios")
plt.tight_layout(); plt.show()
```
