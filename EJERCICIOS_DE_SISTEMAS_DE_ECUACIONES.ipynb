EJERCICIO 01:

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------
# Ejercicio 1a
# Sistema: 5x + 2y = 1 ; -3x + 3y = 5
# ----------------------
A1a = np.array([[5, 2], [-3, 3]])
B1a = np.array([1, 5])
sol1a = np.linalg.solve(A1a, B1a)

x_sol, y_sol = sol1a
print(f"Ejercicio 1a - Soluci√≥n: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

x_vals = np.linspace(-5, 5, 200)
y1a = (1 - 5*x_vals)/2
y2a = (5 + 3*x_vals)/3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1a, label="5x + 2y = 1")
plt.plot(x_vals, y2a, label="-3x + 3y = 5")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 1a")
plt.show()


# ----------------------
# Ejercicio 1b
# Sistema: 2x + y = 6 ; 4x + 3y = 14
# ----------------------
A1b = np.array([[2, 1], [4, 3]])
B1b = np.array([6, 14])
sol1b = np.linalg.solve(A1b, B1b)

x_sol, y_sol = sol1b
print(f"Ejercicio 1b - Soluci√≥n: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

y1b = 6 - 2*x_vals
y2b = (14 - 4*x_vals)/3

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1b, label="2x + y = 6")
plt.plot(x_vals, y2b, label="4x + 3y = 14")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 1b")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ----------------------
# Ejercicio 4a
# Sistema: -2x + 3y = 14 ; 3x - y = -14
# ----------------------
A4a = np.array([[-2, 3], [3, -1]])
B4a = np.array([14, -14])
sol4a = np.linalg.solve(A4a, B4a)

x_sol, y_sol = sol4a
print(f"Ejercicio 4a - Soluci√≥n: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")

x_vals = np.linspace(-10, 10, 200)
y1a = (14 + 2*x_vals)/3
y2a = 3*x_vals + 14

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1a, label="-2x + 3y = 14")
plt.plot(x_vals, y2a, label="3x - y = -14")
plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 4a")
plt.show()


# ----------------------
# Ejercicio 4b
# Sistema: 2x + 3y = 2 ; -6x + 12y = 1
# ----------------------
A4b = np.array([[2, 3], [-6, 12]])
B4b = np.array([2, 1])

try:
    sol4b = np.linalg.solve(A4b, B4b)
    x_sol, y_sol = sol4b
    print(f"Ejercicio 4b - Soluci√≥n: x = {x_sol:.2f} (eje X), y = {y_sol:.2f} (eje Y)")
except np.linalg.LinAlgError:
    print("Ejercicio 4b - El sistema no tiene soluci√≥n √∫nica (es incompatible o tiene infinitas soluciones).")
    sol4b = None

y1b = (2 - 2*x_vals)/3
y2b = (1 + 6*x_vals)/12

plt.figure(figsize=(6,6))
plt.plot(x_vals, y1b, label="2x + 3y = 2")
plt.plot(x_vals, y2b, label="-6x + 12y = 1")
if sol4b is not None:
    plt.scatter(x_sol, y_sol, color="red", zorder=5, label=f"Sol ({x_sol:.2f}, {y_sol:.2f})")
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.legend()
plt.grid(True)
plt.title("Ejercicio 4b")
plt.show()
```

```python
# Ejercicio 19
# a) Obt√©n tres puntos de la recta ax + by = c (sustituye los puntos por los que veas en el gr√°fico)
# b) Halla tres soluciones de la ecuaci√≥n ax + by = c (ser√°n los mismos puntos)
# c) Explica la relaci√≥n entre puntos de la recta y soluciones

import numpy as np
import matplotlib.pyplot as plt

# --- REEMPLAZA estos puntos por los 3 puntos que leas en la gr√°fica del enunciado ---
points_from_graph = [(0, 1), (2, 2.3), (6, 4)]  # <-- Ejemplo: cambia por los reales del gr√°fico
# ------------------------------------------------------------------------------

# Tomamos los dos primeros puntos para obtener la pendiente / ecuaci√≥n
(x1, y1), (x2, y2), (x3, y3) = points_from_graph

# Calcular la recta en forma y = m x + b
if x2 == x1:
    raise ValueError("Los dos primeros puntos verticales: escoge otros dos con diferente x.")
m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

# Pasar a forma ax + by = c  -->  m x - y + b = 0  =>  a = m, b_coeff = -1, c = -b
a = m
b_coeff = -1
c = -b

# Normalizar para enteros (opcional) ‚Äî aqu√≠ lo hacemos para presentar coeficientes m√°s limpios
from fractions import Fraction
def approx_rational(val, max_denom=100):
    return Fraction(val).limit_denominator(max_denom)

a_r = approx_rational(a)
b_r = approx_rational(b_coeff)
c_r = approx_rational(c)

print("Ecuaci√≥n (forma y = m x + b):")
print(f"  y = {m:.6f} x + {b:.6f}")
print("Ecuaci√≥n (forma ax + by = c):")
print(f"  {a_r} x + ({b_r}) y = {c_r}   (racional aproximada)")

# Verificar que los 3 puntos est√°n en la recta
print("\nVerificaci√≥n de puntos (deben cumplir ax + by = c):")
for px, py in points_from_graph:
    lhs = a*px + b_coeff*py
    print(f" Point ({px}, {py}): LHS = {lhs:.6f}, RHS = {c:.6f} -> cumple: {abs(lhs-c) < 1e-6}")

# Mostrar tres soluciones expl√≠citas (usamos los mismos puntos le√≠dos)
solutions = points_from_graph[:3]
print("\nTres soluciones (ejemplos):")
for s in solutions:
    print(" ", s)

# Grafico: recta + puntos
x_vals = np.linspace(min(px for px,_ in points_from_graph)-2, max(px for px,_ in points_from_graph)+2, 400)
y_vals = m * x_vals + b

plt.figure(figsize=(7,6))
plt.plot(x_vals, y_vals, label=f"Recta: y = {m:.3f}x + {b:.3f}")
# marcar los 3 puntos
for i,(px,py) in enumerate(points_from_graph):
    plt.scatter(px, py, zorder=5, label=f"P{i+1} {px,py}")
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Ejercicio 19 ‚Äî Recta y puntos le√≠dos")
plt.show()

# Explicaci√≥n:
print("\nRespuesta (c):")
print(" Todos los puntos situados sobre la recta son soluciones de la ecuaci√≥n ax + by = c.")
print(" Es decir, si (x,y) est√° en la recta, entonces sustituyendo en ax+by=c se cumple la igualdad.")
```

```python
# Ejercicio 20
# a) De los pares: (0,10), (3,19/2), (-1,-4), (0,2/5), (-1/2,7)
#    ¬øcu√°les son soluciones de -3x + (1/2) y = 5 ?
# b) Representa gr√°ficamente la recta -3x + (1/2) y = 5.
# c) ¬øQu√© relaci√≥n hay entre los puntos de la recta y las soluciones?

import numpy as np
import matplotlib.pyplot as plt
from fractions import Fraction

# Lista de pares (usar Fraction para conservar exactitud si se desea)
pairs = [
    (0, 10),
    (Fraction(3,1), Fraction(19,2)),   # (3, 19/2)
    (-1, -4),
    (0, Fraction(2,5)),
    (Fraction(-1,2), 7)
]

# Ecuaci√≥n: -3x + (1/2)y = 5  -> despejar y: (1/2)y = 5 + 3x  -> y = 2*(5 + 3x) = 10 + 6x
def lhs_value(x, y):
    return -3*x + Fraction(1,2)*y

print("Comprobaci√≥n de pares respecto a -3x + (1/2)y = 5:\n")
solutions_found = []
for p in pairs:
    x_p, y_p = p
    val = lhs_value(x_p, y_p)
    is_solution = (val == 5)
    print(f" Punto {p}: LHS = {val} -> soluci√≥n: {is_solution}")
    if is_solution:
        solutions_found.append(p)

print("\nPuntos que son soluciones:", solutions_found)

# Graficar la recta y marcar puntos
x_vals = np.linspace(-6, 6, 400)
y_vals = 10 + 6*x_vals  # como mostramos arriba: y = 10 + 6x

plt.figure(figsize=(7,6))
plt.plot(x_vals, y_vals, label=r"-3x + (1/2)y = 5  (=> y = 10 + 6x)")

# marcar todos los pares; si cumplen, color verde, sino color red
for p in pairs:
    xp = float(p[0])
    yp = float(p[1])
    color = 'green' if p in solutions_found else 'red'
    plt.scatter(xp, yp, color=color, zorder=5)
    plt.text(xp+0.1, yp+0.1, str(p), fontsize=9)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Ejercicio 20 ‚Äî Recta y comprobaci√≥n de pares")
plt.show()

# Respuesta (c)
print("\nRespuesta (c):")
print(" Los puntos que est√°n sobre la recta (es decir, los que al sustituir en la ecuaci√≥n dan igualdad) son exactamente las soluciones.")
print(" Cada punto (x,y) que yace en la recta satisface -3x + (1/2)y = 5; los dem√°s no.")
```

```python
# Problema 12
# Mezclamos dos l√≠quidos:
# Precio1 = 0.94 ‚Ç¨/L, Precio2 = 0.86 ‚Ç¨/L
# Obtenemos 40 L de mezcla a 0.89 ‚Ç¨/L.
# ¬øCu√°ntos litros de cada tipo usamos?
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x, y = sp.symbols('x y', real=True)

# Interpretaci√≥n:
# x + y = 40
# 0.94*x + 0.86*y = 0.89 * 40
eq1 = sp.Eq(x + y, 40)
eq2 = sp.Eq(sp.Rational(94,100)*x + sp.Rational(86,100)*y, sp.Rational(89,100)*40)

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 12 - Resultado")
print(f"  Litros del primer l√≠quido (0.94 ‚Ç¨/L): {x_val:.2f} L")
print(f"  Litros del segundo l√≠quido (0.86 ‚Ç¨/L): {y_val:.2f} L")
print(f"  Comprobaci√≥n precio medio = {(0.94*x_val + 0.86*y_val)/40:.3f} ‚Ç¨/L")

# Gr√°fico: pastel de proporci√≥n
plt.figure(figsize=(5,4))
plt.pie([x_val, y_val], labels=[f"0.94 ‚Ç¨/L\n{x_val:.1f} L", f"0.86 ‚Ç¨/L\n{y_val:.1f} L"], autopct='%1.1f%%', colors=['#66c2a5','#fc8d62'])
plt.title("Problema 12 ‚Äî Proporci√≥n de la mezcla (40 L)")
plt.show()
```

**Problema n¬∫ 13**

```python
# Problema 13
# "El doble de un n√∫mero m√°s la mitad de otro suman 7;
#  si sumamos 7 al primero, obtenemos el qu√≠ntuple del otro."
# Plantea y resuelve el sistema.

import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

x, y = sp.symbols('x y', real=True)

# Interpretaci√≥n:
# 2x + (1/2) y = 7
# x + 7 = 5*y
eq1 = sp.Eq(2*x + sp.Rational(1,2)*y, 7)
eq2 = sp.Eq(x + 7, 5*y)

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 13 - Resultado")
print(f"  Primer n√∫mero x = {x_val:.3f}")
print(f"  Segundo n√∫mero y = {y_val:.3f}")

# Graficaci√≥n: las dos rectas en el plano (x,y) y su intersecci√≥n
x_vals = np.linspace(x_val-5, x_val+5, 400)
# despejar y de cada ecuaci√≥n para graficar
y_from_eq1 = [float((7 - 2*xx) * 2) for xx in x_vals]     # from eq1: (1/2)y = 7 - 2x => y = 2*(7 - 2x)
y_from_eq2 = [float((xx + 7)/5) for xx in x_vals]         # eq2: y = (x + 7)/5

plt.figure(figsize=(6,5))
plt.plot(x_vals, y_from_eq1, label="2x + 1/2 y = 7  -> y = 2(7-2x)")
plt.plot(x_vals, y_from_eq2, label="x + 7 = 5y       -> y = (x+7)/5")
plt.scatter([x_val], [y_val], color='red', label=f"Soluci√≥n ({x_val:.3f},{y_val:.3f})")
plt.xlabel("x"); plt.ylabel("y")
plt.legend(); plt.grid(True); plt.title("Problema 13 ‚Äî Intersecci√≥n de condiciones")
plt.show()
```

**Problema n¬∫ 15**

```python
# Problema 15
# Inversi√≥n 1: beneficio 5%   -> 0.05 * x
# Inversi√≥n 2: beneficio 3.5% -> 0.035 * y
# Se invirti√≥ en total 10000 ‚Ç¨: x + y = 10000
# Los beneficios de la primera inversi√≥n superan en 300 ‚Ç¨ a los de la segunda:
# 0.05 x - 0.035 y = 300
# Hallar cu√°nto se invirti√≥ en cada producto.

import sympy as sp
import matplotlib.pyplot as plt

x, y = sp.symbols('x y', real=True)

eq1 = sp.Eq(x + y, 10000)
eq2 = sp.Eq(sp.Rational(5,100)*x - sp.Rational(35,1000)*y, 300)  # 0.05x - 0.035y = 300

sol = sp.solve((eq1, eq2), (x, y), dict=True)[0]
x_val = float(sol[x])
y_val = float(sol[y])

print("Problema 15 - Resultado")
print(f"  Inversi√≥n en primer producto (5%)  = {x_val:.2f} ‚Ç¨")
print(f"  Inversi√≥n en segundo producto (3.5%)= {y_val:.2f} ‚Ç¨")
print(f"  Beneficio 1 = {0.05*x_val:.2f} ‚Ç¨, Beneficio 2 = {0.035*y_val:.2f} ‚Ç¨")
print(f"  Diferencia de beneficios = {0.05*x_val - 0.035*y_val:.2f} ‚Ç¨ (comprobaci√≥n)")

# Gr√°fico: barras comparando las inversiones y beneficios
plt.figure(figsize=(7,4))
plt.subplot(1,2,1)
plt.bar(['Inv1 (5%)','Inv2 (3.5%)'], [x_val, y_val], color=['#4c72b0','#dd8452'])
plt.title("Inversiones")
plt.ylabel("Euros")
plt.subplot(1,2,2)
plt.bar(['Beneficio1','Beneficio2'], [0.05*x_val, 0.035*y_val], color=['#4c72b0','#dd8452'])
plt.title("Beneficios")
plt.tight_layout(); plt.show()
```
